```python
import numpy as np
from PIL import Image

from yolov7 import YOLOv7, InferenceArgs
```

```python
args = InferenceArgs("weights/yolov7.pt")
args
```

```python
yolo = YOLOv7(args)
```

```python
image = Image.open("inference/images/bus.jpg")
prediction = yolo(np.asarray(image))
prediction
```

```python
visualized_result = Image.fromarray(prediction.visualize())
visualized_result.thumbnail((512, 512))
visualized_result
```
